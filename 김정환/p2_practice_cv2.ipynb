```python
import cv2, json, glob, os
import numpy as np
from matplotlib import pyplot as plt
cv2.__version__
```

```python
#opencv로 전처리 이미지 로드(흑백)

img=cv2.imread("./1.Training/f86f_done.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap="gray"), plt.axis("off")
plt.show()
# type(img)   #데이터 타입 확인
# img         #이미지 데이터 확인
# img.shape   #차원 확인(해상도)
print(type(img), img.shape, sep='\n')
```

```python
#컬러로 이미지 로드
img_bgr=cv2.imread("./1.Training/f86f_done.jpg", cv2.IMREAD_COLOR)
img_bgr[0,0]    #픽셀 확인
img_rgb=cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)    #rgb로 변환
plt.imshow(img_rgb), plt.axis("off")
plt.show()
```

```python
#이미지 저장
img=cv2.imread("./1.Training/f86f_done.jpg", cv2.IMREAD_GRAYSCALE)
cv2.imwrite("./1.Training/f86f_done_new.jpg", img)
```

```python
#이미지 크기 변경

import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread("./1.Training/f86f_done.jpg", cv2.IMREAD_GRAYSCALE)

img_resized=cv2.resize(img, (64, 64))
cv2.imshow('resize', img_resized)

plt.imshow(img_resized, cmap="gray"), plt.axis("off")
plt.show()
```
